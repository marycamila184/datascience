## Leitura e análise inicial dos dados
Foi iniciado um DataFrame a partir do arquivo 2019-09-01_sigesguarda_-_Base_de_Dados.csv.

```python
#importação de bibliotecas
import pandas as pd

%matplotlib inline

# lê o arquivo CSV da guarda municipal
df = pd.read_csv('data/2019-10-01_sigesguarda_-_Base_de_Dados.csv', encoding='latin-1', sep=';', low_memory=False)
```

Algumas linhas para verficar se tudo foi carregado corretamente



```python
df.head(10)
```

Método info para exibir as colunas e os tipos identificados.

```python
df.info()
```

Analisando o dicionario de dados do df percebi que não vou utilizar todas as colunas pois ha informação duplicada e também há muita informação na natureza e na subcategoria. Com isto optei por gerar outro df somente com colunas com informação da data da ocorrência, bairro da ocorrência, qual é a principal natureza da ocorrência, qual é a natureza secundaria da ocorrência, qual é a principal subcategoria da ocorrência, qual é a subcategoria secundária, qual serviço que originou o chamado e a regional em que o chamado ocorreu.

```python
df = df[['ATENDIMENTO_BAIRRO_NOME', 'NATUREZA1_DESCRICAO', 'NATUREZA2_DESCRICAO', 'SUBCATEGORIA1_DESCRICAO', 'SUBCATEGORIA2_DESCRICAO', 'OCORRENCIA_DATA', 'ORIGEM_CHAMADO_DESCRICAO', 'REGIONAL_FATO_NOME']]
df.head()
```

## Localizando e tratando valores inválidos
Checo a quantidade de dados nulos em todas as colunas do df

```python
df[df.isna().any(axis=1)]
```

```python
df.isnull().sum()
```

Verifico que ha muita informação nula em NATUREZA2_DESCRICAO e SUBCATEGORIA2_DESCRICAO entao opto por retirar estas duas colunas. Ja na coluna SUBCATEGORIA1_DESCRICAO também há muita informação nula porém de acordo com o dicionário de dados esta coluna é descrição da sub-categoria da primeira natureza então resolvo continuar utilizando-a porém vou tratar os dados nulos colocando a string 'Não informado'

```python
# Retirando 'NATUREZA2_DESCRICAO' e 'SUBCATEGORIA2_DESCRICAO'
df = df[['ATENDIMENTO_BAIRRO_NOME', 'NATUREZA1_DESCRICAO', 'SUBCATEGORIA1_DESCRICAO', 'OCORRENCIA_DATA', 'ORIGEM_CHAMADO_DESCRICAO', 'REGIONAL_FATO_NOME']]
df.head()
```

```python
# Colocando o valor 'Não informado' na coluna SUBCATEGORIA1_DESCRICAO
df['SUBCATEGORIA1_DESCRICAO'] = df['SUBCATEGORIA1_DESCRICAO'].fillna('Não informado')
df.head(25)
```

Novamente faço a checagem de dados nulos nas colunas

```python
df[df.isna().any(axis=1)]
```

```python
df.isnull().sum()
```

Porém agora opto em remover os dados que possuem ATENDIMENTO_BAIRRO_NOME nulo pois são poucas entradas e sem os dados do bairro de atendimento nao tenho como analisar de onde veio esta ocorrência.

```python
df = df.dropna(subset=["ATENDIMENTO_BAIRRO_NOME"])
df[df.isna().any(axis=1)]
```

```python
df.isnull().sum()
```

Fazendo a contagem novamente percebemos que as linhas foram removidas e que todos os registro nulos da coluna REGIONAL_FATO_NOME também foram removidos também. Agora vou tratar as entredas inválidas, começo com o comando unique para ver os dados não repetidos

```python
df["ATENDIMENTO_BAIRRO_NOME"].unique()
```

Olhando os valores da coluna bairro opto por retirar todos os valores que contém 'INDICAÇÕES CANCELADA','BAIRRO NAO INFORMADO', 'BAIRRO FICTÍCIO','NÃO ENCONTRADO','SEM DADOS','BAIRRO NÃO LOCALIZAD','NÃO INFORMADO ','NI','NF'.

```python
df = df[df["ATENDIMENTO_BAIRRO_NOME"] != 'INDICAÇÕES CANCELADA']
df = df[df["ATENDIMENTO_BAIRRO_NOME"] != 'BAIRRO NAO INFORMADO']
df = df[df["ATENDIMENTO_BAIRRO_NOME"] != 'BAIRRO FICTÍCIO']
df = df[df["ATENDIMENTO_BAIRRO_NOME"] != 'NÃO ENCONTRADO']
df = df[df["ATENDIMENTO_BAIRRO_NOME"] != 'SEM DADOS']
df = df[df["ATENDIMENTO_BAIRRO_NOME"] != 'BAIRRO NÃO LOCALIZAD']
df = df[df["ATENDIMENTO_BAIRRO_NOME"] != 'NÃO INFORMADO ']
df = df[df["ATENDIMENTO_BAIRRO_NOME"] != 'NI']
df = df[df["ATENDIMENTO_BAIRRO_NOME"] != 'NF']

df["ATENDIMENTO_BAIRRO_NOME"].unique()
```

Observando o resultado do método unique percebemos que há bairros duplicados então começo o tratamento das strings

```python
df['ATENDIMENTO_BAIRRO_NOME'] = df['ATENDIMENTO_BAIRRO_NOME'].str.upper()
df['ATENDIMENTO_BAIRRO_NOME'] = df['ATENDIMENTO_BAIRRO_NOME'].str.replace('JD.','JARDIM')
df['ATENDIMENTO_BAIRRO_NOME'] = df['ATENDIMENTO_BAIRRO_NOME'].str.replace('JD','JARDIM')
df['ATENDIMENTO_BAIRRO_NOME'] = df['ATENDIMENTO_BAIRRO_NOME'].str.replace('LOT.','LOTEAMENTO')

df["ATENDIMENTO_BAIRRO_NOME"].unique()
```

Mesmo assim ainda temos valores inválidos como JARDIMIPE, CAMPO DE SÃO BENEDIT, SÃO THOMAS e SÃO THOMAZ então faço uma pesquisa para verificar o nome correto dos bairros para aplicar o método replace

```python
df['ATENDIMENTO_BAIRRO_NOME'] = df['ATENDIMENTO_BAIRRO_NOME'].str.replace('JARDIMEUROPA','JARDIM EUROPA')
df['ATENDIMENTO_BAIRRO_NOME'] = df['ATENDIMENTO_BAIRRO_NOME'].str.replace('JARDIMSUISSA','JARDIM SUISSA')
df['ATENDIMENTO_BAIRRO_NOME'] = df['ATENDIMENTO_BAIRRO_NOME'].str.replace('VILA PERNETA ','VILA PERNETA')
df['ATENDIMENTO_BAIRRO_NOME'] = df['ATENDIMENTO_BAIRRO_NOME'].str.replace('CAMPO DE SÃO BENEDIT','CAMPO DE SÃO BENEDITO')
df['ATENDIMENTO_BAIRRO_NOME'] = df['ATENDIMENTO_BAIRRO_NOME'].str.replace('MARIA ANTONIETA','VILA MARIA ANTONIETA')
df['ATENDIMENTO_BAIRRO_NOME'] = df['ATENDIMENTO_BAIRRO_NOME'].str.replace('SÃO THOMAS','SÃO THOMAZ')
df['ATENDIMENTO_BAIRRO_NOME'] = df['ATENDIMENTO_BAIRRO_NOME'].str.replace('JARDIMIPE','JARDIM IPE')
df['ATENDIMENTO_BAIRRO_NOME'] = df['ATENDIMENTO_BAIRRO_NOME'].str.replace('AGUAS BELAS','ÁGUAS BELAS')
df['ATENDIMENTO_BAIRRO_NOME'] = df['ATENDIMENTO_BAIRRO_NOME'].str.replace('CIC','CIDADE INDUSTRIAL')
df['ATENDIMENTO_BAIRRO_NOME'] = df['ATENDIMENTO_BAIRRO_NOME'].str.replace('THOMAS COELHO','THOMAZ COELHO')
df['ATENDIMENTO_BAIRRO_NOME'] = df['ATENDIMENTO_BAIRRO_NOME'].str.replace('IGUAÇÚ 1','IGUAÇU 01')

df["ATENDIMENTO_BAIRRO_NOME"].unique()
```

Agora vou tratar os dados da coluna REGIONAL_FATO_NOME que tambem e bairro

```python
df["REGIONAL_FATO_NOME"].unique()
```

Opto por deixar os valores em caixa alta e trocar CIC por CIDADE INDUSTRIAL

```python
df['REGIONAL_FATO_NOME'] = df['REGIONAL_FATO_NOME'].str.upper()
df['REGIONAL_FATO_NOME'] = df['REGIONAL_FATO_NOME'].str.replace("CIC", "CIDADE INDUSTRIAL")
df["REGIONAL_FATO_NOME"].unique()
```

Agora vou tratar os dados da coluna ORIGEM_CHAMADO_DESCRICAO e percebo que não há informação inválida ou dupicada porém opto por retirar os parentese do CIOSP (190) e deixar somente com 190.

```python
df["ORIGEM_CHAMADO_DESCRICAO"].unique()
```

```python
df['ORIGEM_CHAMADO_DESCRICAO'] = df['ORIGEM_CHAMADO_DESCRICAO'].str.replace(r"\(.*\)","")
df['ORIGEM_CHAMADO_DESCRICAO'] = df['ORIGEM_CHAMADO_DESCRICAO'].str.replace("CIOSP ","190")

df["ORIGEM_CHAMADO_DESCRICAO"].unique()
```

Começo o tratamento da coluna SUBCATEGORIA1_DESCRICAO. Segundo o dicionário de dados esta coluna e uma descrição da natureza entçao opto somente por aplicar o método lower em algumas strings e retirar os parentese padronizano com um hifen. Em alguns casos retiro a informação dentro do parênteses pois não será importante para a análise.

```python
df["SUBCATEGORIA1_DESCRICAO"].unique()
```

```python
df['SUBCATEGORIA1_DESCRICAO'] = df['SUBCATEGORIA1_DESCRICAO'].str.replace(r"\(.*\)","")
df['SUBCATEGORIA1_DESCRICAO'] = df['SUBCATEGORIA1_DESCRICAO'].str.replace("PRESTAÇÃO DE SOCORRO/SALVAMENTO", "Socorro")
df['SUBCATEGORIA1_DESCRICAO'] = df['SUBCATEGORIA1_DESCRICAO'].str.replace("ORIENTAÇÃO", "Orientação")
df['SUBCATEGORIA1_DESCRICAO'] = df['SUBCATEGORIA1_DESCRICAO'].str.replace("Disparo de Alarme ", "Disparo de Alarme")

df["SUBCATEGORIA1_DESCRICAO"].unique()
```

Começo o tratamento da coluna NATUREZA1_DESCRICAO. Opto por tirar os parentese e os conteudos pois nao sera importante para a análise.

```python
df["NATUREZA1_DESCRICAO"].unique()
```

```python
df['NATUREZA1_DESCRICAO'] = df['NATUREZA1_DESCRICAO'].str.replace(r"\(.*\)","")

df['NATUREZA1_DESCRICAO'].unique()
```

## Conversão de tipos, criação de coluna e escrita em CSV

Exibo o df e os tipos das colunas do df novamente

```python
df.info()
df.head(10)
```

Converto os valores das colunas que contém bairros e a descrição da ocorrência para caixa baixa. Opto por não converter a coluna ORIGEM_CHAMADO_DESCRICAO pois temos siglas de órgãos municipais e também temos números como origem.

```python
df["ATENDIMENTO_BAIRRO_NOME"] = df["ATENDIMENTO_BAIRRO_NOME"].str.lower()
df["REGIONAL_FATO_NOME"] = df["REGIONAL_FATO_NOME"].str.lower()
df["NATUREZA1_DESCRICAO"] = df["NATUREZA1_DESCRICAO"].str.lower()
df["SUBCATEGORIA1_DESCRICAO"] = df["SUBCATEGORIA1_DESCRICAO"].str.lower()
df.head(10)
```

Converto as colunas para o tipo string usando o método astype e converto a coluna de data para datetime.

```python
df = df.astype({'ATENDIMENTO_BAIRRO_NOME':str, 'NATUREZA1_DESCRICAO':str, 'SUBCATEGORIA1_DESCRICAO': str, 'ORIGEM_CHAMADO_DESCRICAO':str, 'REGIONAL_FATO_NOME':str  })
df["OCORRENCIA_DATA"] = pd.to_datetime(df["OCORRENCIA_DATA"])

df.info()
```

Exporto o df para um csv para fazer a análise exploratória

```python
df.to_csv('data/2019-10-01_sigesguarda_-_Base_de_Dados-limpo.csv', index=False)
```
